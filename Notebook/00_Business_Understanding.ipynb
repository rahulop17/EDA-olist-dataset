# 🛒 Olist Brazilian E-Commerce Business Analytics

## Project Overview

This project analyzes the Brazilian E-Commerce Public Dataset provided by Olist to uncover actionable business insights. The analysis focuses on understanding sales performance, customer behavior, product performance, seller efficiency, delivery operations, and customer satisfaction.

The goal is to perform an end-to-end Exploratory Data Analysis (EDA) using Python and SQL to answer key business questions that can support data-driven decision making.

---

## Objectives

- Understand the structure and quality of the dataset
- Perform data cleaning and preprocessing
- Engineer meaningful business features
- Analyze customer purchasing behavior
- Evaluate seller and product performance
- Investigate delivery efficiency
- Identify factors affecting customer satisfaction
- Generate actionable business recommendations

# Business Problem

Olist is a Brazilian e-commerce marketplace that connects thousands of sellers with customers across Brazil.

As the platform grows, leadership wants to better understand customer behavior, operational performance, seller efficiency, and overall business growth.

This analysis aims to answer business questions that will help improve customer satisfaction, optimize logistics, identify high-performing products and sellers, and support strategic business decisions.

# Business Questions

This project aims to answer the following business questions:

## Business Growth
1. How has revenue changed over time?
2. How many orders are placed each month?
3. What is the average order value?

## Customer Analysis
4. Which states generate the highest revenue?
5. Which cities have the most customers?
6. What percentage of customers are repeat buyers?
7. Who are the high-value customers?

## Product Analysis
8. Which product categories generate the highest revenue?
9. Which categories receive the lowest ratings?
10. Which categories perform well in sales but poorly in customer satisfaction?

## Delivery Analysis
11. What is the average delivery time?
12. Which states experience the longest delivery times?
13. Do delivery delays affect customer review scores?

## Seller Analysis
14. Which sellers generate the highest revenue?
15. Which sellers provide the fastest delivery?
16. Which sellers receive the highest and lowest customer ratings?

## Payment Analysis
17. Which payment methods are most commonly used?
18. How do customers prefer to pay for their purchases?

## Final Goal
Provide actionable business recommendations based on the analysis.
