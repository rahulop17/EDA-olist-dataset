# Data Loading and Understanding

In this notebook, we will:

- Import the required libraries
- Load all datasets
- Explore the structure of each dataset
- Check data types
- Check missing values
- Understand the relationship between datasets

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)
```

```python
customers = pd.read_csv("../Data/olist_customers_dataset.csv")
orders = pd.read_csv("../Data/olist_orders_dataset.csv")
order_items = pd.read_csv("../Data/olist_order_items_dataset.csv")
payments = pd.read_csv("../Data/olist_order_payments_dataset.csv")
reviews = pd.read_csv("../Data/olist_order_reviews_dataset.csv")
products = pd.read_csv("../Data/olist_products_dataset.csv")
sellers = pd.read_csv("../Data/olist_sellers_dataset.csv")
geolocation = pd.read_csv("../Data/olist_geolocation_dataset.csv")
category_translation = pd.read_csv("../Data/product_category_name_translation.csv")
```

```python
datasets = {
    "Customers": customers,
    "Orders": orders,
    "Order Items": order_items,
    "Payments": payments,
    "Reviews": reviews,
    "Products": products,
    "Sellers": sellers,
    "Geolocation": geolocation,
    "Category Translation": category_translation
}

for name, df in datasets.items():
    print(f"{name:<25} {df.shape}")
```

#Dataset information

```python
for name, df in datasets.items():
    print("=" * 80)
    print(name.upper())
    print("=" * 80)
    df.info()
    print()
```

# Missing Values

```python
for name, df in datasets.items():
    print("=" * 80)
    print(name.upper())
    print("=" * 80)
    print(df.isnull().sum())
    print()
```

# Duplicate Records

```python
for name, df in datasets.items():
    print(f"{name:<25} {df.duplicated().sum()}")
```

# Dataset Dimensions

```python
summary = pd.DataFrame(columns=["Rows", "Columns"])

for name, df in datasets.items():
    summary.loc[name] = [df.shape[0], df.shape[1]]

summary
```

# Column Names

```python
for name, df in datasets.items():
    print("="*70)
    print(name.upper())
    print("="*70)
    print(df.columns.tolist())
    print()
```

# Unique Values

```python
for name, df in datasets.items():
    print("="*70)
    print(name.upper())
    print("="*70)
    print(df.nunique())
    print()
```

# Memory Usage

```python
for name, df in datasets.items():
    memory = df.memory_usage(deep=True).sum() / (1024**2)
    print(f"{name:<25} {memory:.2f} MB")
```

## Key Findings

- The dataset consists of 9 relational tables.
- The Geolocation dataset contains duplicate records.
- Missing values are primarily present in the Orders, Reviews, and Products datasets.
- Most datasets have complete primary key information.
- Date columns are currently stored as object data types and will require conversion during data cleaning.
